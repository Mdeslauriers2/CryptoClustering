```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import holoviews as hv
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
#Code from ChatGPT
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
#From ChatGPT
crypto = pd.DataFrame(scaled_data, columns=df_market_data.columns, index=df_market_data.index)

# Copy the crypto names from the original data
#From ChatGPT
crypto_names = df_market_data.index.copy()
crypto_names_df = pd.DataFrame(crypto_names, columns=['coin_id'])

# Set the coinid column as index
"""Coin id is already index in the original df"""
# Display sample data
crypto.head(5)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(crypto)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve

elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
crypto_elbow_curve=pd.DataFrame(elbow_data)
crypto_elbow_curve.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
crypto_elbow_curve.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=2, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
model.fit(crypto)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
grouped_crypto = model.predict(crypto)

# Print the resulting array of cluster values.
print(grouped_crypto)
```

```python
# Create a copy of the DataFrame
crypto_predictions_df = crypto.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
crypto_predictions_df['prediction'] = grouped_crypto

# Display sample data
crypto_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
crypto_predictions_df.hvplot.scatter(
    x="price_change_percentage_24h", 
    y="price_change_percentage_7d", 
    #by="customer rating",
    hover_cols='coin_id'
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_info = pca.fit_transform(crypto)

# View the first five rows of the DataFrame. 
pca_info[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:**  0.8949

```python
# Creating a DataFrame with the PCA data
pca_info = pd.DataFrame(
    pca_info,
    columns=["PCA1", "PCA2", "PCA3"],
    index=crypto_names_df['coin_id']  # Set the 'coin_id' column as the index

)
# Copy the crypto names from the original data
crypto_names = df_market_data.index.copy()
crypto_names_df = pd.DataFrame(crypto_names, columns=['coin_id'])

# Set the coinid column as index
"""coin_id is the index in the original dataset"""

# Display sample data
pca_info.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data)
    inertia.append(k_model.inertia_)
    
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following questions:

  * **Answer:** 4

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
model.fit(pca_info)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
k_3 = model.predict(pca_info)

# Print the resulting array of cluster values.
k_3
```

```python
# Create a copy of the DataFrame with the PCA data
cryptoinfo_pca_predictions_df = pca_info.copy()


# Add a new column to the DataFrame with the predicted clusters
cryptoinfo_pca_predictions_df["predicted_clusters"] = k_3

# Display sample data
cryptoinfo_pca_predictions_df.head(5)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
cryptoinfo_pca_predictions_df.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="predicted_clusters",  # Adjusted column name
    hover_cols='coin_id'
)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
original_elbow_curve = hv.Curve(df_elbow, 'k', 'inertia', label='Original Data')
pca_elbow_curve = hv.Curve(crypto_elbow_curve, 'k', 'inertia', label='PCA Data')

# Overlay the two curves
composite_plot = original_elbow_curve * pca_elbow_curve

# Customize the plot
composite_plot.opts(
    title='Elbow Curve Comparison',
    xlabel='Number of Clusters (k)',
    ylabel='Inertia',
    legend_position='top_left'
)
```

```python
# Composite plot to contrast the clusters
# Cluster plot for original data
original_cluster_plot = hv.Scatter(crypto_predictions_df, 'price_change_percentage_24h', 'price_change_percentage_7d', group='prediction', label='Original Data')

# Cluster plot for PCA-transformed data
pca_cluster_plot = hv.Scatter(cryptoinfo_pca_predictions_df, 'PCA1', 'PCA2', group='predicted_clusters', label='PCA Data').opts(color='red')

# Overlay the two cluster plots
composite_plot = original_cluster_plot * pca_cluster_plot

# Customize the plot
composite_plot.opts(
    title='Cluster Plot Comparison',
    xlabel='Feature 1',
    ylabel='Feature 2',
    legend_position='top_left'
)
```
